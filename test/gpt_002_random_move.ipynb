```python
import matplotlib.pyplot as plt
import random

# Define the starting point
x, y = 0, 0

# Path history
x_positions = [x]
y_positions = [y]

# Move right until (4, 0)
while x < 4:
    x += 1
    x_positions.append(x)
    y_positions.append(y)

# Probabilistic decision at (4, 0)
if random.random() < 0.5:
    # Move to (5, 1) and stop
    x, y = 5, 1
    x_positions.append(x)
    y_positions.append(y)
else:
    # Continue moving right to (10, 0)
    while x < 10:
        x += 1
        x_positions.append(x)
        y_positions.append(y)

# Plot the movement
plt.figure(figsize=(8, 6))
plt.plot(x_positions, y_positions, marker="o", markersize=5, label="Path")
plt.scatter([x_positions[0]], [y_positions[0]], c="green", label="Start")
plt.scatter([x_positions[-1]], [y_positions[-1]], c="red", label="End")
plt.title("Probabilistic Point Movement")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(4, color="blue", linestyle="--", linewidth=0.5, label="Decision Point")
plt.legend()
plt.grid(True)
plt.axis("equal")
plt.show()
```
