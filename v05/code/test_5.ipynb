```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
# Your data: background and binary overlay
background = np.random.rand(6, 6) * 100
overlay = (np.random.rand(6, 6) > 0.5).astype(int)

fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(background, cmap='Blues', alpha=0.7)

# Annotate only where overlay = 1
for i in range(overlay.shape[0]):
    for j in range(overlay.shape[1]):
        if overlay[i, j] == 1:
            # Show overlay value in center
            ax.text(j, i, f'{overlay[i, j]}',
                   ha='center', va='center',
                   color='red', fontweight='bold', fontsize=12)
        
        else:
            # Optional: show background value (smaller, less prominent)
            ax.text(j, i, f'{background[i, j]:.1f}',
                   ha='center', va='center',
                   color='gray', fontsize=8, alpha=0.7)
        
        

plt.colorbar(im, label='Background Values')
plt.title('Overlay (1) shown in red circles\nBackground values in gray')
plt.show()
```
