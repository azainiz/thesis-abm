```python
import matplotlib.pyplot as plt
import numpy as np

# Create sample data
data1 = np.array([[1, 0, 1, 0],
                  [0, 1, 0, 1],
                  [1, 1, 0, 0],
                  [0, 0, 1, 1]])

data2 = np.array([[0.2, 0.8, 0.3, 0.7],
                  [0.6, 0.4, 0.9, 0.1],
                  [0.3, 0.5, 0.7, 0.2],
                  [0.8, 0.6, 0.4, 0.9]])

fig, ax = plt.subplots(figsize=(6, 6))

# Plot data2 as background (grayscale)
im2 = ax.imshow(data2, cmap='gray', alpha=0.7)

# Plot data1 as overlay (colored)
im1 = ax.imshow(data1, cmap='Reds', alpha=0.5)

plt.colorbar(im1, label='Data 1')
plt.colorbar(im2, ax=ax, label='Data 2')
plt.title('Overlay with Alpha Transparency')
plt.show()
```
