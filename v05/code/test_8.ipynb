```python
import pandas as pd
import random

# Generate random x,y pairs
data = []
for i in range(4):
    x = random.random()  # Random float between 0 and 1
    y = random.random()
    data.append([x, y])

# Create DataFrame
df = pd.DataFrame(data, columns=['rest area visitors', 'what is this'])

# Save to CSV
df.to_csv('random_data.csv', index=False, float_format='%.6f')

print("Saved 4 random pairs to random_data.csv")
print(df)
```
