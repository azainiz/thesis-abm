```python

x = []

# Count occurrences
count_1 = x.count(1)
count_2 = x.count(2)
count_3 = x.count(3)

# How many 1s should be in total
target_1_count = count_2 + count_3

# How many more 1s to add
to_add = target_1_count - count_1

# Add the required number of 1s
x.extend([1] * to_add)

print(x)
```
